```python
data.groupby('location')['sql_marks'].sum().plot(kind='bar')
plt.title('Total sql_marks by location')
plt.xlabel('location')
plt.ylabel('sql_marks')
plt.show()
data.groupby('location')['excel_marks'].sum().plot(kind='bar')
plt.title('Total excel_marks by location')
plt.xlabel('location')
plt.ylabel('excel_marks')
plt.show()
data.groupby('location')['python_marks'].sum().plot(kind='bar')
plt.title('Total python_marks by location')
plt.xlabel('location')
plt.ylabel('python_marks')
plt.show()
data.groupby('location')['power_bi_marks'].sum().plot(kind='bar')
plt.title('Total power_bi_marks by location')
plt.xlabel('location')
plt.ylabel('power_bi_marks')
plt.show()
data.groupby('location')['english_marks'].sum().plot(kind='bar')
plt.title('Total english_marks by location')
plt.xlabel('location')
plt.ylabel('english_marks')
plt.show()
```
